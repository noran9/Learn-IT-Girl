```python
import numpy as np
import pandas as pd
import requests
import json

url = "http://api.apixu.com/v1/history.json?key=f87b86729047441786b83603190105&q=Gursuf, Krym, Ukraine&dt=2019-04-25"
JSONContent = requests.get(url).json()
content = json.dumps(JSONContent, indent = 4, sort_keys=True)
print(content)
```

```python
print(JSONContent['forecast']['forecastday'][0]['hour'][0]['temp_c'])
print(JSONContent['forecast']['forecastday'][0]['hour'][0]['pressure_mb'])
print(JSONContent['forecast']['forecastday'][0]['hour'][0]['precip_mm'])
print(JSONContent['forecast']['forecastday'][0]['hour'][0]['humidity'])
print(JSONContent['forecast']['forecastday'][0]['hour'][0]['cloud'])
```

```python
cities = ['Skopje', 'London', 'Ljubljana', 'Paris', 'Rome', 'Berlin', 'Madrid', 'Barcelona', 'Stockholm', 'Sofia', 'Belgrade', 'Graz']
cities_list = []

for city in cities:
    for i in range(25, 30):
        JSONContent = requests.get("http://api.apixu.com/v1/history.json?key=f87b86729047441786b83603190105&dt=2019-04-" + str(i) + "&q=" + city).json()
        for j in range(0, 23):
            cities_list.append([city, JSONContent['forecast']['forecastday'][0]['hour'][j]['temp_c'], 
                              JSONContent['forecast']['forecastday'][0]['hour'][j]['pressure_mb'],
                              JSONContent['forecast']['forecastday'][0]['hour'][j]['humidity'], 
                              JSONContent['forecast']['forecastday'][0]['hour'][j]['cloud']])

dataset = pd.DataFrame(cities_list)
```

```python
dataset.columns = ["City", "Temperature", "Pressure", "Humidity", "Cloud cover"]
dataset.tail()
```
